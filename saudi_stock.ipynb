```python
saudi_stock['target_class']=(saudi_stock['target']>0).astype(int)

saudi_stock_ml = pd.get_dummies(saudi_stock, columns=['sector', 'day_of_week'], drop_first=True)

print("Target distribution:")
print(saudi_stock_ml['target_class'].value_counts(normalize=True))
```

```python
saudi_stock_ml = saudi_stock_ml.sort_values(['symbol', 'date']).reset_index(drop=True)

# Create 1-day and 2-day lags for key financial indicators
for lag in [1, 2]:
    saudi_stock_ml[f'return_lag_{lag}'] = saudi_stock_ml.groupby('symbol')['return'].shift(lag)
    saudi_stock_ml[f'volume_lag_{lag}'] = saudi_stock_ml.groupby('symbol')['volume_traded'].shift(lag)
    saudi_stock_ml[f'spread_lag_{lag}'] = saudi_stock_ml.groupby('symbol')['Spread'].shift(lag)
    saudi_stock_ml[f'trades_lag_{lag}'] = saudi_stock_ml.groupby('symbol')['no_trades'].shift(lag)


lag_cols = [c for c in saudi_stock_ml.columns if 'lag' in c]
print("Generated lag features:", lag_cols)
```

```python

saudi_stock_ml['close_rolling_mean_5'] = saudi_stock_ml.groupby('symbol')['close'].transform(lambda x: x.rolling(5).mean())
saudi_stock_ml['close_rolling_mean_20'] = saudi_stock_ml.groupby('symbol')['close'].transform(lambda x: x.rolling(20).mean())
saudi_stock_ml['momentum_ratio'] = saudi_stock_ml['close_rolling_mean_5'] / saudi_stock_ml['close_rolling_mean_20']


saudi_stock_ml['volatility_10d'] = saudi_stock_ml.groupby('symbol')['return'].transform(lambda x: x.rolling(10).std())

# 3. Volume Breakdown (Is volume exploding compared to its 15-day average?)
saudi_stock_ml['volume_velocity'] = saudi_stock_ml['volume_traded'] / saudi_stock_ml.groupby('symbol')['volume_traded'].transform(lambda x: x.rolling(15).mean())
```

```python
# Drop rows with NaN values resulting from lags and targets
saudi_stock_ml = saudi_stock_ml.dropna(subset=lag_cols + ['target_class']).reset_index(drop=True)

# Define feature matrix (X) and label vector (y)
# Dynamically collect all one-hot encoded sector/day features and lag features
encoded_features = [col for col in saudi_stock_ml.columns if 'sector_' in col or 'day_of_week_' in col]
base_features = ['year', 'month', 'Spread'] + lag_cols

X_features = base_features + encoded_features
X = saudi_stock_ml[X_features]
y = saudi_stock_ml['target_class']

print(f"Total rows for training: {X.shape[0]} | Number of features: {X.shape[1]}")
```

```python
import numpy as np

# 1. Replace infinite values with NaN
X = X.replace([np.inf, -np.inf], np.nan)

# 2. Check if we have any NaNs now (including the ones from inf)
print("NaNs found per feature after converting inf:")
print(X.isnull().sum())

# 3. Align y with X by dropping rows where X has NaNs
valid_indices = X.dropna().index
X = X.loc[valid_indices].reset_index(drop=True)
y = y.loc[valid_indices].reset_index(drop=True)

# Also ensure our split mask data frames align properly
saudi_stock_ml_cleaned = saudi_stock_ml.loc[valid_indices].reset_index(drop=True)
```

```python
train_mask = saudi_stock_ml['date'] < '2025-01-01'
test_mask = saudi_stock_ml['date'] >= '2025-01-01'

X_train, y_train = X[train_mask], y[train_mask]
X_test, y_test = X[test_mask], y[test_mask]

print(f"Train samples: {X_train.shape[0]} (Pre-2025)")
print(f"Test samples: {X_test.shape[0]} (2025-2026 Out-of-Sample)")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix


lr_model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
lr_model.fit(X_train, y_train)
lr_preds = lr_model.predict(X_test)


rf_model = RandomForestClassifier(n_estimators=150, max_depth=10, class_weight='balanced', random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)

print("--- Logistic Regression Baseline Results ---")
print(f"Accuracy: {accuracy_score(y_test, lr_preds):.4f}")
print(classification_report(y_test, lr_preds))

print("\n--- Random Forest Classifier Results ---")
print(f"Accuracy: {accuracy_score(y_test, rf_preds):.4f}")
print(classification_report(y_test, rf_preds))
```

```python

from xgboost import XGBClassifier

# Calculate a precise scale weight for XGBoost based on your data distribution
# (approx 2443 negative rows / 1731 positive rows)
ratio = y_train.value_counts()[0] / y_train.value_counts()[1]


xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    scale_pos_weight=ratio,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)

print("--- XGBoost Classifier Results ---")
print(f"Accuracy: {accuracy_score(y_test, xgb_preds):.4f}")
print(classification_report(y_test, xgb_preds))
```
